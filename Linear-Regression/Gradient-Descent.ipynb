```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
X = np.arange(10)
Y = (X-5)**2
```

```python
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('f(X)')
plt.show()
```

```python
x = 0
lr = 0.1
err = []

plt.plot(X,Y,'g')
for i in range(100):
    g = 2*x - 10
    

    x = x - lr*g
    
    
    y = (x-5)**2 
    
    err.append(y)
    plt.scatter(x,y,c='red')
    
    print(x)
    
    
```
