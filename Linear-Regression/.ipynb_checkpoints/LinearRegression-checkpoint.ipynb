```python
def estimated_salary(m,x,c):
  return m * x + c
```

```python
m = 100
c = 500
```

```python
x = data['Years of Experience']
y = data['Salary']
estimated_salary(m,x,c)
```

```python
plt.plot(x,estimated_salary(m,x,c),'g-o')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

```python
plt.plot(x,estimated_salary(m,x,c),'g-o')
plt.scatter(x,y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['Estimated','Actual'])
plt.show()
```

```python
def try_params(m,c):
  plt.plot(x,estimated_salary(m,x,c),'g-o')
  plt.scatter(x,y)
  plt.xlabel('Years of Experience')
  plt.ylabel('Salary')
  plt.legend(['Estimated','Actual'])
  plt.show()
```

```python
try_params(600,1000)
```

```python
def loss(y,y_pred):
  rmse = np.sqrt(np.mean(np.square(y-y_pred)))
  return rmse
```

```python
y_pred = estimated_salary(1000,x,1000)
```

```python
loss(y,y_pred)
```

```python
def best_params():
  my_list = []
  m = 1
  c = 1
  for i in range(1,70000):
    es_sal = estimated_salary(m,x,c)
    loss1 = loss(y,es_sal)
    my_list.append([loss1,m,c])
    

    m = m + 0.63
    c = c + 0.36554
  
  return sorted(my_list)[0]
    

```

```python
best_params()
```

```python
import statsmodels.api as sm
```

```python
x = data['Years of Experience']
y = data['Salary']
```

```python
x
```

```python
x1 = sm.add_constant(x)
```

```python
x1.head()
```

```python
model = sm.OLS(y,x1).fit()
```

```python
model
```

```python
model.summary()
```

```python
model.params
```

```python
m = model.params[1]
c = model.params[0]
```

```python
loss(y,estimated_salary(m,x,c))
```

```python
plt.scatter(x,y)
yhat = m * x + c
fig = plt.plot(x,yhat,c='red',label = 'regression line',lw = 4) 
plt.xlabel('Years of Experience',fontsize = 20)
plt.ylabel('Salary', fontsize = 20)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR_model = LinearRegression()
```

```python
LR_model.fit(x1,y)
```

```python
LR_model.score(x1,y)
```

```python
LR_model.intercept_
```

```python
LR_model.coef_[1]
```

```python
loss(y,estimated_salary(LR_model.coef_[1],x,LR_model.intercept_))
```
