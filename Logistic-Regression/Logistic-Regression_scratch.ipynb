```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
mean1 = np.array([2,1])
cov1 = np.array([[2.3,1.7],[1.7,4.3]])

mean2 = np.array([10,15])
cov2 = np.array([[2.7,2.1],[2.1,3.9]])

dist1 = np.random.multivariate_normal(mean1,cov1,1000)
dist2 = np.random.multivariate_normal(mean2,cov2,1000)
```

```python
data[:1000,:2] = dist1
data[1000:,:2] = dist2
```

```python
data[1000:,-1] = 1.0
```

```python
split = int(0.75*data.shape[0])

X_train = data[:split,:-1]
X_test = data[split:,:-1]

Y_train = data[:split,-1]
Y_test = data[split:,-1]
```

Creating Logistic Regression Model.

```python
def sigmoid(x):
    sig = 1 / (1 + np.exp(-x))
    return sig
```

```python
def pred(X,theta):
    return sigmoid(np.dot(X,theta))
```

```python
def error(X,y,theta):
    y_pred = pred(X,theta)
    err = -1 * np.mean((y * np.log(y_pred)) + ((1 - y) *  np.log(1 - y_pred)))
    return err
```

```python
def gradient(X,y,theta):
    y_pred = pred(X,theta)    
    grad = np.dot(X.T,(y - y_pred))
    m = X.shape[0]        
    return grad / m
```

```python
def gradient_descent(X,y,lr = 0.1,max_steps = 1500):
    n = X.shape[1]
    theta = np.zeros((n,1))
    err_list = []
    
    for i in range(max_steps):
        err = error(X,y,theta)
        err_list.append(err)
        grad = gradient(X,y,theta)
        
        theta = theta + (lr * grad)
        
    return (theta,err_list)
        
```

```python
import statsmodels.api as sm
X_train_new = sm.add_constant(X_train)
```
